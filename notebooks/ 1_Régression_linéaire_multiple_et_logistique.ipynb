# Modèle régression linéaire multiple sélection descendante (backward selection)

```python
import statsmodels.api as sm

# Définition des variables (X = toutes sauf quality, Y = quality)
X = df_standardized.drop(columns=['quality'])
Y = df_standardized['quality']

# Ajout de la constante (β0)
X = sm.add_constant(X)

# Ajustement du modèle par la méthode des Moindres Carrés (OLS)
model = sm.OLS(Y, X).fit()

# Affichage du résumé complet (Table d'ANOVA, Coef β, R², Test F)
print(model.summary())
```

Analyse de la Variance (ANOVA) et Test de Fisher :
Pour Prob (F-statistic) on obtient  5,83 x10^-124, ce qui est extrêmement inférieur au seuil de 0,05.
On rejette l'hypothèse nulle H0 : 𝛽1 =𝛽2 =⋯=𝛽𝑝 =0
Au moins une des variables explicatives contribue de manière significative à expliquer la note de qualité
Analyse de la Significativité des Variables (Test de Student) :

L'examen des p-values permet d'identifier les régresseurs réellement influents (ceux dont la p-value est < 0,05)
Les variables significatives avec leur impact via le coefficient beta sont les suivantes :

Impact positif majeur :
*   l’alcohol P=0,000 / coefficient Beta 0,3132
*   les sulphates P=0,000 / coefficient Beta 0,1561

Impact négatif majeur :
*   la volatile acidity P=0,000 / coefficient Beta -0,2050
*   les total sulfur dioxide P=0,001 / coefficient Beta -0,0904
*   les chlorides P=0,000 / coefficient Beta -0,0953

# Sélection du sous-modèle avec les variables significatives

```python
import statsmodels.api as sm

colonnes_optimisees = [
    'alcohol',
    'sulphates',
    'volatile acidity',
    'total sulfur dioxide',
    'pH',
    'chlorides'
]

X_opt = df_standardized[colonnes_optimisees]
Y = df_standardized['quality']

# constante β0  pour estimer l'ordonnée à l'origine
X_opt = sm.add_constant(X_opt)

# Ajustement du nouveau modèle (Moindres Carrés Ordinaires)
model_opt = sm.OLS(Y, X_opt).fit()

# Affichage des résultats
print("Résumé du Sous-Modèle Optimisé ")
print(model_opt.summary())
```

Malgré l'optimisation, plusieurs éléments indiquent que la Régression Linéaire Multiple (OLS) n'est pas l'outil final idéal pour traiter la "qualité" d'un vin car :
*   Faiblesse du R2 qui est seulement  de 36%
*   Pas de  Normalité des résidus : Tests Omnibus et Jarque-Bera tests  qui vérifient si les erreurs suivent une distribution gaussienne (loi normale).
*   Omnibus (Prob: 0.000) et Jarque-Bera (Prob: 1.92e-08), dans les deux cas, la "Probabilité" (p-value) est extrêmement faible (bien inférieure à 0,05).
*   On rejette l'hypothèse nulle de normalité, les erreurs du modèle ne sont pas normalement distribuées









# Confirmation graphique du rejet de l'hypothèse de normalité des résidus


```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as stats
import seaborn as sns

# Calcul des résidus et des valeurs prédites
residus = model_opt.resid
valeurs_predites = model_opt.fittedvalues

#  Création de la figure de diagnostic
plt.figure(figsize=(16, 10))

#  Distribution des résidus (Histogramme)
plt.subplot(2, 2, 1)
sns.histplot(residus, kde=True, color='skyblue')
plt.title('Distribution des Résidus (Vérification de la Normalité)')
plt.xlabel('Résidus')

# QQ-Plot (Normalité théorique)
# Les points doivent suivre la ligne rouge pour confirmer la loi normale
ax2 = plt.subplot(2, 2, 2)
sm.qqplot(residus, line='45', fit=True, ax=ax2)
plt.title('QQ-Plot des Résidus')

# Résidus vs Valeurs Prédites (Homoscédasticité)
# On doit obtenir un nuage de points aléatoire sans forme particulière
plt.subplot(2, 2, 3)
plt.scatter(valeurs_predites, residus, alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Résidus vs Valeurs Prédites')
plt.xlabel('Qualité Prédite')
plt.ylabel('Résidus')

#  Valeurs Réelles vs Valeurs Prédites
# Plus les points sont proches de la ligne, meilleur est le modèle
plt.subplot(2, 2, 4)
plt.scatter(df_standardized['quality'], valeurs_predites, alpha=0.3)
plt.plot([3, 8], [3, 8], color='red', linestyle='--')
plt.title('Qualité Réelle vs Qualité Prédite')
plt.xlabel('Qualité Réelle')
plt.ylabel('Qualité Prédite')

plt.tight_layout()
plt.show()
```



*   L'histogramme montre une distribution des erreurs qui n'est pas parfaitement symétrique.
*   Les points sur le QQ-Plot ne suivent pas parfaitement la ligne droite rouge, surtout aux extrémités.
*   Pour le graphique Résidus vs Qualité Prédites, les résidus ne sont pas distribués de manière aléatoire et uniforme (hétéroscédasticité).
*   Enfin pour le graphique Qualité Réelle vs Prédite, le modèle a un faible pouvoir de discrimination. Il ne peut pas prédire la "catégorie" exacte du vin d'où l'intêret de tester la regression logistique.








# Modèle de régression logistique

Transformation en binaire de la variable quality pour prédire le succès ou l’échec ici la bonne qualité ou mauvaise qualité du vin. Car pour la regression logistique on se place toujours dans le cadre de la classification binaire.

```python
# Création d'une variable binaire "quality_binary"
# 1 = bon (quality >= 6)
# 0 = mauvais (quality <= 5)
df["quality_binary"] = (df["quality"] >= 6).astype(int)

# Vérification
print(df[["quality", "quality_binary"]].head())
print(df["quality_binary"].value_counts())
```

Histogramme Barplot de quality_binary

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="quality_binary")
plt.title("Distribution de quality_binary (0 = mauvais, 1 = bon)")
plt.xlabel("quality_binary")
plt.ylabel("Nombre d'échantillons")
plt.show()
```

Pré-traitement et séparation des données

```python
# Séparation des données : 80% Entraînement, 20% Test
# random_state assure la reproductibilité du découpage
Y = df["quality_binary"]
X = df[
    ['alcohol', 'sulphates', 'volatile acidity', 'total sulfur dioxide', 'pH', 'chlorides']
]

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)

print(f"\nDimension des données d'entraînement : {X_train.shape}")
print(f"Dimension des données de test : {X_test.shape}")
```

Entrainement uniquement sur l'ensemble d'entraînement

```python
X_train_scaled = scaler.fit_transform(X_train)
```

Application de la même transformation à l'ensemble de test

```python
X_test_scaled = scaler.transform(X_test)
```

Données standardisées dans le DataFrame

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

Entraînement du Modèle Initialisation du modèle de Régression Logistique

```python
# random_state pour la reproductibilité
from sklearn.linear_model import LogisticRegression
logreg_model = LogisticRegression(random_state=42)
```

Entraînement du modèle sur les données standardisées

```python
logreg_model.fit(X_train_scaled, Y_train)
```

Interprétation des Coefficients

```python
# Récupération des coefficients (w) et du biais (b)
coefficients = pd.Series(logreg_model.coef_[0], index=X.columns)
bias = logreg_model.intercept_[0]

print(f"Biais (Intercept - b) : {bias:.4f}")
print("Coefficients (poids w) :")
print(coefficients.sort_values(ascending=False).round(4))
```

Facteur influençant positivement la qualité du vin :

*   Alcool (1.0188) : Est la variable la plus influente.
*   Sulphates (0.4711) : Ils contribuent positivement à la qualité.

Facteur de dégradation de la qualité du vin :

*   Volatile Acidity (-0.4453) et total sulfur dioxide (-0,4096) sont les facteurs principaux qui diminuent fortement la qualité du vin.


```python
# Interprétation des coefficients (Odds Ratio)
print(" Analyse des Odds Ratios (Impact)")
# L'exponentielle du coefficient (exp(w)) donne l'Odds Ratio (OR)
# Un OR > 1 signifie que l'augmentation de X augmente l'ODDS (Probabilité) de Y=1.
# Un OR < 1 signifie que l'augmentation de X diminue l'ODDS de Y=1.
odds_ratios = np.exp(coefficients).sort_values(ascending=False).round(2)
print(odds_ratios)
```

Évaluation des Résultats

```python

# Prédictions sur l'ensemble de Test
from sklearn.metrics import accuracy_score, confusion_matrix
Y_pred = logreg_model.predict(X_test_scaled)
Y_proba = logreg_model.predict_proba(X_test_scaled)[:, 1]

print(" Métriques de Performance (Ensemble de Test) ")

# Précision (Accuracy)
accuracy = accuracy_score(Y_test, Y_pred)
print(f"\nPrécision (Accuracy) : {accuracy:.4f}")

from sklearn.metrics import f1_score, classification_report
#Calcul du F1-score simple
f1 = f1_score(Y_test, Y_pred)
print(f"F1-score : {f1:.4f}")
```

L'évaluation finale sur l'ensemble de test valide la pertinence du modèle de régression logistique avec une précision globale de 73,16 % et un F1-score de 0,7439

# Rapport de classification

```python
from sklearn.metrics import classification_report
report = classification_report(Y_test, Y_pred)
print("\nRapport de Classification :")
print(report)
```

Le support indique que j’ai  128 vins "Mauvais" (0) et 144 vins "Bons" (1) dans l’échantillon de test.

Le Recall  est la  métrique qui mesure la capacité d'un modèle à débusquer tous les individus d'une classe donnée.

*   Précision = Vrais positifs/(Vrai Positifs+Faux positifs)

*   Pour les "Bons Vins" (Classe 1) : la précision est de  0.75, cela veut dire que lorsque le modèle prédit qu'un vin est bon, il a raison dans 75 % des cas. Le modèle parvient à détecter 74 % de tous les bons vins.
*   Pour les "Mauvais Vins" (Classe 0) Le modèle est légèrement moins précis  il a raison dans 71 % des cas. Le modèle parvient à détecter 73 % de tous les mauvais vins.






```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc
import numpy as np
import pandas as pd

# Création d'une figure avec 3 sous-graphiques
plt.figure(figsize=(18, 5))

# --- 1. Matrice de Confusion (Heatmap) ---
plt.subplot(1, 3, 1)
cm = confusion_matrix(Y_test, Y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Mauvais (0)', 'Bon (1)'],
            yticklabels=['Mauvais (0)', 'Bon (1)'])
plt.title('1. Matrice de Confusion', fontsize=14)
plt.xlabel('Prédiction')
plt.ylabel('Réel')

# --- 2. Courbe ROC ---
plt.subplot(1, 3, 2)
fpr, tpr, thresholds = roc_curve(Y_test, Y_proba)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de Faux Positifs (FPR)')
plt.ylabel('Taux de Vrais Positifs (TPR)')
plt.title('2. Courbe ROC', fontsize=14)
plt.legend(loc="lower right")

# --- 3. Importance des Variables (Coefficients) ---
plt.subplot(1, 3, 3)
# Récupération des noms de colonnes et des coefficients
feature_names = ['alcohol', 'sulphates', 'volatile acidity', 'total sulfur dioxide', 'pH', 'chlorides']
coefs = logreg_model.coef_[0]
features_importance = pd.Series(coefs, index=feature_names).sort_values()

features_importance.plot(kind='barh', color=['red' if x < 0 else 'green' for x in features_importance])
plt.title('3. Importance des Variables (Coefficients $w$)', fontsize=14)
plt.axvline(x=0, color='black', lw=1)
plt.xlabel('Poids du coefficient')

plt.tight_layout()
plt.show()
```

Courbe ROC :

*   Le score de 0.81 correspond à l'Aire Sous la Courbe (AUC) il donne la capacité globale du modèle à distinguer les bons vins des mauvais,  le score de 0.81 indique que dans 81 % des cas, le modèle classera correctement un vin de qualité au-dessus d'un vin médiocre.

Importance des variables (coefficients w) :


*   Cela confirme que l’alcohol est la variable dominante pour que le vin soit jugé bon à l’opposé l’acidité volatile tend a donner un vin de mauvaise qualité.
